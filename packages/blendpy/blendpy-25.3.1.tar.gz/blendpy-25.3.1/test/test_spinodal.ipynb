```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_spinodal = pd.read_csv("spinodal.csv")
df_binodal = pd.read_csv("binodal.csv")
```

```python
df_AgPt = pd.read_csv("binodal_AgPt.csv")
df_AuPt = pd.read_csv("binodal_AuPt.csv")
```

```python
plt.xlim(0, 1)
plt.ylim(300, 2500)
plt.plot(df_spinodal['x'], df_spinodal['t'])
plt.plot(df_binodal['x'], df_binodal['t'])
plt.show()
```

```python
plt.xlim(0, 1)
plt.ylim(300, 2500)
plt.plot(df_AgPt['x'], df_AgPt['t'], label='AgPt')
plt.plot(df_AuPt['x'], df_AuPt['t'], label='AuPt')
plt.legend()
plt.show()
```
