# Viz
Let's learn how to build visualizations
## Introduction

````python
from zoopy import animal, viz
````

## Functions

```python
plot_classification(animal_for_plotting: animal.Animal) -> None:
```
Plots a hierarchical classification graph

### Parameters
```animal_for_plotting (zoopy.animal.Animal)```: an animal for plot classification

### Examples

````python
penguin = animal.Animal('пингвин', 'ru')
penguin.display()
````

````python
viz.plot_classification(penguin)
````
